```python
# 2. Performing Customers/sellers Segmentation
# a. Divide the customers into groups based on the revenue generated

customer_group=data_model.groupby("customer_id")["payment_value"].sum().reset_index()
customer_group["Revenue_group"]=pd.qcut(customer_group["payment_value"],q=3,labels=['Low Revenue', 'Medium Revenue', 'High Revenue'])
customer_group
```

```python
# b. Divide the sellers into groups based on the revenue generated

seller_group=data_model.groupby("seller_id")["payment_value"].sum().reset_index()
seller_group["Revenue_group"]=pd.qcut(seller_group["payment_value"],q=3,labels=['Low Revenue', 'Medium Revenue', 'High Revenue'])
seller_group
```

```python
# 3. Cross-Selling (Which products are selling together)
# Hint: We need to find which of the top 10 combinations of products are selling together in
# each transaction. (combination of 2 or 3 buying together)
from itertools import combinations
from collections import Counter
combination_counter = Counter()

order_products=data_model.groupby("order_id")["product_id"].apply(list)
for products in order_products:
    # Generate all combinations of 2 and 3 products
    product_combinations = [comb for i in range(2, 4) for comb in combinations(products, i)]
    # Update the counter with the combinations
    combination_counter.update(product_combinations)

# Get the top 10 combinations
top_10_combinations = combination_counter.most_common(10)

print("Top 10 combinations of products selling together:")
for combination, frequency in top_10_combinations:
    print(combination, ": Frequency =", frequency)
```
